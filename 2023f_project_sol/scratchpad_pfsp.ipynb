```python
# Εκτύπωση της πρώτης γραμμής κάθε αρχείου προβλήματος
for fn in sorted(glob.glob('./Taillard-PFSP/ta*')):
    print(fn)
    with open(fn, 'r') as f:
        print(f.readline())
```

```python
# Ανάγνωση περιεχομένων αρχείου προβλήματος, τοποθέτηση σε μια δομή (αριθμός εργασιών, αριθμός μηχανών, λεξικό με χρόνους επεξεργασίας)
def read_pfsp_instance(fn):
    with open(fn, 'r') as f:
        n_jobs, n_machines = map(int, f.readline().split())
        p = {}
        for j in range(n_jobs):
            for i, t in enumerate(f.readline().split()[1::2]):
                p[j,i]=int(t)
        return n_jobs, n_machines, p

n_jobs, n_machines, p = read_pfsp_instance('./Taillard-PFSP/ta001')
n_jobs, n_machines, p
```

```python
# κλήση της συνάρτησης read_pfsp_instance για το πρόβλημα toy_5j_3m
n_jobs, n_machines, p = read_pfsp_instance('./toy_5j_3m')
n_jobs, n_machines, p
```

```python
import numpy as np

# ένας απλοϊκός επιλυτής
def naive_solution(n_jobs, seed=None):
    if seed is not None:
        np.random.seed(seed)
    return np.random.permutation(n_jobs)

# δημιουργία πίνακα χρόνων τερματισμού εκτέλεσης κάθε εργασίας σε κάθε μηχάνημα
def schedule(n_jobs, n_machines, p, solution):
    C = np.zeros((n_jobs, n_machines))
    for idx, j in enumerate(solution):
        for i in range(n_machines):
            if idx == 0:
                if i==0:
                    C[j,i] = p[j,i]
                else:
                    C[j,i] = C[j,i-1] + p[j,i]
            else:
                if i == 0:
                    C[j,i] = C[solution[idx-1],i] + p[j,i]
                else:
                    C[j,i] = max(C[j,i-1], C[solution[idx-1],i]) + p[j,i]
    return C

# χρόνος ολοκλήρωσης προγράμματος
def makespan(job_sequence, C):
    return C[job_sequence[-1], -1]

n_jobs, n_machines, p = read_pfsp_instance('./toy_5j_3m')
# seed=42 για να έχουμε πάντα το ίδιο αποτέλεσμα
job_sequence = naive_solution(n_jobs, seed=42)
print(job_sequence)
C = schedule(n_jobs, n_machines, p, job_sequence)
C
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def display_schedule(C):
    n_jobs, n_machines = C.shape
    colors = list(mcolors.TABLEAU_COLORS.values()) 

    plt.figure(figsize=(10, n_machines * 3))
    for i in range(n_machines):
        for j in job_sequence:
            plt.barh(i, width=p[j, i], left=C[j,i] - p[j, i], height=0.8, color=colors[j % len(colors)], label=f"Job {j}" if i==0 else "")
            
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.yticks(range(n_machines), [f"Machine {i}" for i in range(n_machines)])
    plt.xlabel('Time')
    plt.title('Permutation Flowshop Schedule')
    plt.legend(by_label.values(), by_label.keys(), loc="upper right")   
    plt.gca().invert_yaxis() 
    plt.show()

n_jobs, n_machines, p = read_pfsp_instance('./toy_5j_3m')
job_sequence = naive_solution(n_jobs)
C = schedule(n_jobs, n_machines, p, job_sequence)
print(makespan(job_sequence, C))
display_schedule(C)
```
